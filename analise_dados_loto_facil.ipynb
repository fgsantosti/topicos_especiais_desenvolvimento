# Lotofácil

```python
# Carregar o arquivo CSV contendo os resultados da Lotofácil
import pandas as pd
loto_df = pd.read_csv("/content/loto_facil.csv")
display(loto_df.head())
```

```python
#print("Colunas do DataFrame:", loto_df.columns.tolist())
#print("Primeiras linhas:")
#print(loto_df.head())
```

```python
# Verificar a frequência de cada número sorteado
numeros = list(range(1, 26))  # Números de 1 a 25

# Inicializar o dicionário de frequências
frequencias = {}
for numero in numeros:
    frequencias[numero] = 0
```

```python
# Contar a frequência de cada número
# As colunas das bolas são: 'bola 1', 'bola 2', ..., 'bola 15'
colunas_bolas = [f'bola {i}' for i in range(1, 16)]
# Contar a frequência de cada número
for index, row in loto_df.iterrows():
    # Para cada sorteio, contar os números que saíram
    for coluna in colunas_bolas:
        numero_sorteado = row[coluna]
        if numero_sorteado in frequencias:
            frequencias[numero_sorteado] += 1
```

```python
# Exibir as frequências
'''
print("Frequência de cada número sorteado:")
for numero, frequencia in frequencias.items():
    print(f"Número {numero}: {frequencia} vezes")
'''
```

```python
# Exibir as frequências ordenadas
frequencias = dict(sorted(frequencias.items(), key=lambda item: item[1], reverse=True))
print("\nFrequência de cada número sorteado (ordenado):")
for numero, frequencia in frequencias.items():
    print(f"Número {numero}: {frequencia} vezes")
```

```python
#Exibir os 15 números mais sorteados
print("\nOs 15 números mais sorteados:")
for i, (numero, frequencia) in enumerate(frequencias.items()):
    if i < 15:
        print(f"Número {numero}: {frequencia} vezes")
    else:
        break
```

```python
#As combinacoes mais frequentes
loto_df['combinacao'] = loto_df[colunas_bolas].apply(lambda row: '-'.join(map(str, sorted(row))), axis=1)
combinacoes_frequentes = loto_df['combinacao'].value_counts().head(10)
print("\nAs 10 combinações mais frequentes:")
for combinacao, frequencia in combinacoes_frequentes.items():
    print(f"Combinação {combinacao}: {frequencia} vezes")
```
