```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### 1. Distribuição de Idade por Sexo
Este histograma mostra a distribuição das idades dos passageiros, separada por sexo. `hue='sex'` diferencia as barras para homens e mulheres, e `multiple='stack'` empilha-as para mostrar a composição total. `kde=True` adiciona uma estimativa da densidade do kernel para suavizar as distribuições.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x="age", hue="sex", multiple="stack", kde=True)
plt.title("Distribution of Age by Sex")
plt.show()
```

### 2. Contagem de Passageiros por Classe
Este gráfico de contagem (`countplot`) exibe o número de passageiros em cada classe (`pclass`), também segregado por sexo (`hue='sex'`). É útil para ver a demografia das classes no Titanic.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=titanic, x="pclass", hue="sex")
plt.title("Passenger Count by Class and Sex")
plt.show()
```

### 3. Taxa de Sobrevivência por Sexo
Um gráfico de barras (`barplot`) que ilustra a taxa média de sobrevivência (`survived`) para homens e mulheres. A altura da barra representa a proporção de sobreviventes dentro de cada grupo.

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=titanic, x="sex", y="survived", palette="viridis")
plt.title("Survival Rate by Sex")
plt.ylabel("Survival Rate")
plt.show()
```

### 4. Taxa de Sobrevivência por Classe de Passageiro
Este `barplot` detalha a taxa de sobrevivência em cada classe de passageiro (`pclass`), novamente dividida por sexo (`hue='sex'`). Permite comparar as taxas de sobrevivência entre classes e gêneros.

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=titanic, x="pclass", y="survived", hue="sex", palette="magma")
plt.title("Survival Rate by Passenger Class and Sex")
plt.ylabel("Survival Rate")
plt.show()
```

### 5. Distribuição da Tarifa
Um histograma que exibe a distribuição das tarifas pagas pelos passageiros (`fare`). `kde=True` também é usado aqui para mostrar a estimativa da densidade da distribuição.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x="fare", kde=True)
plt.title("Distribution of Fare")
plt.show()
```

### 6. Idade vs. Tarifa, Colorido pela Sobrevivência
Este gráfico de dispersão (`scatterplot`) compara a idade (`age`) com a tarifa (`fare`), e os pontos são coloridos de acordo com o status de sobrevivência (`hue='survived'`). Isso pode revelar padrões sobre como idade, tarifa e sobrevivência estão relacionados.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic, x="age", y="fare", hue="survived", palette="coolwarm", alpha=0.7)
plt.title("Age vs Fare, Colored by Survival")
plt.show()
```
