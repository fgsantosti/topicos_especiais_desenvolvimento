```python
import pandas as pd
```

```python
df1 = pd.read_csv("PARTICIPANTES_2024.csv", sep = ";", encoding = "ISO-8859-1")
```

```python
df1.head()
```

```python
quantidade_raca = df1["TP_COR_RACA"]
```

```python
#Quantidade de pessoas por raça
quantidade_raca.value_counts()
```

```python
#Quantidade de treineiros
treineiros = df1["IN_TREINEIRO"]
print(treineiros.count())
```

```python
treineiros.value_counts()
```

```python
#Quantidade por sexo
sexo = df1["TP_SEXO"]
sexo.value_counts()
```

```python
#A quantidade por faixa etaria
faixa_etaria = df1["TP_FAIXA_ETARIA"]
faixa_etaria.value_counts()
```

```python
#Quantidade de participantes por estado
estado = df1["SG_UF_PROVA"]
estado.value_counts()
```

```python
#Quantidade de alunos que fizeram a prova no município de Corrente
municipios = df1["NO_MUNICIPIO_PROVA"]
municipio_corrente = municipios[municipios == "Corrente"].count()
print(municipio_corrente)
```
